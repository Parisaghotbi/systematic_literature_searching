```python
###https://medium.com/@darshankhandelwal12/scrape-google-scholar-using-python-3f35a3a6597b
import requests
from bs4 import BeautifulSoup
```

```python
def getScholarData():
    try:
#url = "https://www.google.com/scholar?q=polylactic+acid+pla+film&hl=en"
        url = "https://www.google.com/scholar?q=tensile&hl=en"
        
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        scholar_results = []

        for el in soup.select(".gs_ri"):
            scholar_results.append({
                "title": el.select(".gs_rt")[0].text,
                "title_link": el.select(".gs_rt a")[0]["href"],
                "id": el.select(".gs_rt a")[0]["id"],
                "displayed_link": el.select(".gs_a")[0].text,
                "snippet": el.select(".gs_rs")[0].text.replace("\n", ""),
                "cited_by_count": el.select(".gs_nph+ a")[0].text,
                "cited_link": "https://scholar.google.com" + el.select(".gs_nph+ a")[0]["href"],
                "versions_count": el.select("a~ a+ .gs_nph")[0].text,
                "versions_link": "https://scholar.google.com" + el.select("a~ a+ .gs_nph")[0]["href"] if el.select("a~ a+ .gs_nph")[0].text else "",
            })
      
           
        for i in range(len(scholar_results)):
            scholar_results[i] = {key: value for key, value in scholar_results[i].items() if value != "" and value is not None}

        print(scholar_results)

    except Exception as e:
        print(e)

getScholarData() #nonetype objects
```

```python
# Split the 'column_name' into three new columns
# Define a custom function to handle splitting
df = pd.read_csv('combined_file.csv')
def split_row(row):
    parts = row.split('- ', 2)  # Split into at most 3 parts
    while len(parts) < 3:  # Ensure there are always 3 parts
        parts.append(None)  # Append None for missing values
    return pd.Series(parts)

# Apply the custom function to the 'column_name' column
df[['Author', 'Journalyear', 'Publisher']] = df['displayed_link'].apply(split_row)

# Display the modified DataFrame
print(df)
```

```python
# Remove all duplicate rows based on all columns
df_unique_all_columns = df.drop_duplicates()
print(df_unique_all_columns)
```

```python
df=df_unique_all_columns
df['Journalyear'] = df['Journalyear'].str.strip()

# Adjusted regular expression to extract the year, allowing for potential non-digit characters after the year
df['Year'] = df['Journalyear'].str.extract(r'(\d{4})(?=\D*$)')

# Display the modified DataFrame
print(df)
```

```python
df['Journal'] = df['Journalyear'].str.replace(", \d{4}", "", regex=True)
print(df)
```

```python
# Function to extract number using split
import re

# Function to extract number using regular expressions
def extract_number_regex(text):
    numbers = re.findall(r'\d+', text)
    return int(numbers[0]) if numbers else None
    
# Apply the function to the column
df['Cited_Number'] = df['cited_by_count'].apply(extract_number_regex)
```

```python
df['Cited_Number'] = pd.to_numeric(df['Cited_Number'], errors='coerce')
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

df['CitationNumPerYear']=df['Cited_Number']/(2024-df['Year'])
df
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt

# Ensure you have the necessary NLTK assets
nltk.download('punkt')
nltk.download('stopwords')

# List of titles
titles = pd.array(df['title'])

# Combine all titles into one large string
all_titles = ' '.join(titles).lower()

# Tokenize the string into words
tokens = word_tokenize(all_titles)

# Synonym mapping: Map 'poly' and 'pla' to 'poly_pla'
synonym_map = {'polylactic': 'pla', 'lactic': 'pla', 'poly':'pla', 'films':'film','rabbit':'rabbits','effect':'effects'
               ,'biodegradability': 'biodegradable','essential':'anise','oil':'anise'}
tokens = [synonym_map.get(word, word) for word in tokens]

# Words to omit
words_to_omit = ['html','acid','quality','properties','packaging','applications',
                 'development','application','characterization','review','application','preparation'
                ,'pdf','different','comparison','analysis','effects', 'growth','materials']  # Add more words as needed

# Remove stopwords and punctuation, and filter out the words to omit
tokens = [word for word in tokens if word.isalpha()]  # Remove non-alphabetic tokens
tokens = [word for word in tokens if word not in stopwords.words('english') and word not in words_to_omit]

# Count word frequencies
word_freq = Counter(tokens)

# Display the 10 most common words
most_common_words = word_freq.most_common(20)
print(most_common_words)

# Plot a bar chart
words, counts = zip(*most_common_words)

plt.figure(figsize=(10, 6))  # Create a new figure
plt.bar(words, counts)  # Create a bar plot
plt.title('Most Common Words in Titles')  # Set the title
plt.xlabel('Words')  # Set the x-axis label
plt.ylabel('Frequency')  # Set the y-axis label

# Rotate x-axis labels to vertical
plt.xticks(rotation='vertical')
# Save the plot as a vector graphic
plt.savefig('plot_vector_output.pdf', format='pdf')  # Save as PDF
plt.savefig('plot_vector_output.svg', format='svg')  # Save as SVG

plt.show()
```
